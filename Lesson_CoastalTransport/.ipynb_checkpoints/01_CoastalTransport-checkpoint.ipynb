# Ekman Transport on a Coastline

Here we will explore what happens when we have Ekman transport along a coastline with observations

### Quick Ekman Transport Review

<img src="../Images/ekman.jpg" width="500" align="right">

Recall when wind blows over the surface of the Ocean, this force combined with Coriois effect results in surface water transport to the right of the wind direction. There is in fact a spiral of water transport through depth, but if you average this transport over the surface layer you see an average transport to the right (in the Northern Hemisphere).

### The mooring array

<img src="../Images/EA_WAOR_2018_labels.png" width="500" align="right">

The coastal array contains three moorings. We will be examining observations from the mooring closest to the shore, the inshore mooring (4).

At mooring sites, an instrument package including a CTD (conductivity-temperature-depth) moves up and down continuously sampling the temperature, salinity, and depth profile of the water column. These measurements are then a function of both water depth and time, allowing us to see how properties in the water column change as wind forcing varies.

### Adding wind 

You see above that surface temperatures, and even down to 25-meters appear to change some what rapidly in the summer during August and September. Let's see if this might be related to the wind forcing. Below we will plot both the temperature data and wind data with a little more control over the plotting. Specifically we will make 2-panels, use quiver() to make wind arrows, and restrict our time axis to just August and March.

```python
# First create wind speed variable, quiver() will use this to color the arrows by speed
ds['wind_speed'] = np.sqrt(np.square(ds['wind_u'])+np.square(ds['wind_v']))

# Set up plotting for two figures
f = plt.figure(num=None, figsize=(12,6), dpi=80, facecolor='w', edgecolor='k')
(ax1, ax2) = f.subplots(2, 1, sharex=True)

# Plot first the wind using quiver()
plt.sca(ax1)
time = ds['time'][:] # Need to convert time to datetime64 for quiver, otherwise error
time = np.array(time,dtype=np.datetime64)
Q = plt.quiver(time,0, ds['wind_u'], ds['wind_v'], ds['wind_speed'])
plt.quiverkey(Q, 0.85, 0.8, 5, r'5 m/s', labelpos='E',coordinates='figure')
plt.grid()

# Plot second the temperature
plt.sca(ax2)
ds['temperature'].plot(add_colorbar=False)
plt.gca().invert_yaxis() #<- flips yaxis
#Here let's restrict our time frame to a continuous section of data
ax2.set_xlim([datetime.date(2016, 8, 1), datetime.date(2016, 10, 1)]) 
```

### Downwelling

<img src="../Images/downwelling.jpg" width="500" align="right">

There is a particuarly dramatic change around September 1st, when winds blow strongly from the South. This will result in downwelling, let us see what happens in this case. 

First wind blowing from the South will create a transport of surface waters to the right, or to the East. This means surface waters will run into the coastline. As surface waters continue to be transported East they will pile up, forcing the original water there downwards.

This "downwelled" water will be trasnported both to deeper depths and ultimately westward out to sea. The result is we have a piling up of warmer surface waters at the coast. It looks like this makes sense with what we see in the data, with much warmer temperatures at the coastline as soon as the wind switched from the North-West to the South-East.
